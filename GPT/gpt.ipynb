```python
import os
import pandas as pd
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()


client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),  # This is the default and can be omitted
)

def ask_gpt(question, model):
    response = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": question,
            }
        ],
        model=model,
    )
    return response.choices[0].message.content
```

```python
data_path = "../DATA/questions_by_gpt.csv"

# upload the csv file
csv_path = data_path
df = pd.read_csv(csv_path)
df

# create a list to store the answers of gpt
gpt_answers = []

# get question  in dataframe
list_question = df["Question"].tolist()

# loop through the questions and ask gpt
for question in list_question:
	answer = ask_gpt(question + " The answer must be between 50 and 100 characters", "gpt-3.5-turbo")
	gpt_answers.append(answer)

# add the answers to the dataframe
df["GPT Answer"] = gpt_answers

# save the dataframe to a csv file
df.to_csv(data_path, index=False)
```

```python
data_path = "../DATA/questions.json"

df = pd.read_json(data_path)

questions = df["Question"].tolist()

gpt3_answers = []
gpt4_answers = []

for question in questions:
	answer = ask_gpt(question + " The answer must be between 50 and 100 characters", "gpt-3.5-turbo")
	gpt3_answers.append(answer)
	answer = ask_gpt(question + " The answer must be between 50 and 100 characters", "gpt-4")
	gpt4_answers.append(answer)

# add the answers to the dataframe
df["GPT3 Answer"] = gpt3_answers
df["GPT4 Answer"] = gpt4_answers

# save the dataframe to a json file
df.to_json(data_path, orient="records")
df

```

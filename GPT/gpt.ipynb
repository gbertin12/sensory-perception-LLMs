```python
import os
import pandas as pd
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

data_path = "../DATA/questions_by_human.csv"
#data_path = "../DATA/questions_by_gpt.csv"


client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),  # This is the default and can be omitted
)

def ask_gpt(question):
    response = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": question,
            }
        ],
        model="gpt-4o",
    )
    return response.choices[0].message.content
```

```python
# upload the csv file
csv_path = data_path
df = pd.read_csv(csv_path)
df
```

```python
# create a list to store the answers of gpt
gpt_answers = []

# get question  in dataframe
list_question = df["Question"].tolist()

# loop through the questions and ask gpt
for question in list_question:
	answer = ask_gpt(question + " The answer must be between 50 and 100 characters")
	gpt_answers.append(answer)

gpt_answers
```

```python
# add the answers to the dataframe
df["GPT Answer"] = gpt_answers
df

# save the dataframe to a csv file
df.to_csv(data_path, index=False)
```

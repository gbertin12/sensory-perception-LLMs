```python
import pandas as pd
from mistralai import Mistral
import time
import os
from dotenv import load_dotenv

load_dotenv()

data_path = "../DATA/questions_by_human.csv"
#data_path = "../DATA/questions_by_gpt.csv"

api_key = os.getenv("MISTRAL_API_KEY")

model = "mistral-large-latest"

client = Mistral(api_key=api_key)

# Function to ask Mistral
def ask_mistral(question):
	response = client.chat.complete(
		model=model,
		messages=[
			{
				"role": "user",
				"content": question,
			},
		]
	)
	return response.choices[0].message.content
```

```python
# upload the csv file
csv_path = data_path
df = pd.read_csv(csv_path)
df
```

```python
# create a list to store the answers of Mistral
mistral_answers = []

# get question  in dataframe
list_question = df["Question"].tolist()

# loop through the questions and ask Mistral
for question in list_question:
	answer = ask_mistral(question + " The answer must be between 50 and 100 characters")
	mistral_answers.append(answer)
	time.sleep(10)

mistral_answers
```

```python

# add the answers to the dataframe
df["Mistral Answer"] = mistral_answers
df

# save the dataframe to a csv file
df.to_csv(data_path, index=False)
```
